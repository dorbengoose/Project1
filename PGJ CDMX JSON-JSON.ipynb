* TLAHUAC

```python
from itertools import islice

tlahuac=[]
for i, row in df.iterrows(): #head(500)
    
    try:
        if df["fields"][i]["alcaldia_hechos"]=="TLAHUAC" and df["fields"][i]["ao_hechos"] >= "2016":
            tlahuac.append((df["fields"][i]["delito"],df["fields"][i]["ao_hechos"]))
    except:
        tlahuac.append(np.nan)
```

```python
tlahuacdf=pd.DataFrame(tlahuac)
```

```python
tlahuacdf=tlahuacdf.rename(columns={0:'CRIME', 1:"TLAHUAC"})
tlahuacplot=tlahuacdf.groupby(["CRIME"]).count().sort_values("TLAHUAC", ascending=False).head(20).reset_index(inplace=False)
tlahuacplot
```

```python
from functools import reduce
```

```python
crimelist = [tlahuacplot,iztapalaplot,cuauhtemocplot,venustianoplot,gustavoplot,azcapotzalcoplot,coyoacanplot,miguelplot,tlalpanplot,
          cuajimalpaplot,xochimilcoplot,alvaroplot,benitoplot,magdalenaplot,milpaplot,iztacalcoplot]
```

```python
crimefinal= reduce(lambda left,right: pd.merge(left,right,on='CRIME'), crimelist)
crimefinal.index=crimefinal.index+1
crimefinal
```

```python
crimefinal2=crimefinal.transpose()
crimefinal2.columns = crimefinal2.iloc[0]
crimefinal2.drop(['CRIME'],inplace=True)
crimefinal2.columns = crimefinal2.columns.get_level_values(0)
```

```python

import seaborn as sns
plt.figure(figsize = (12,7))
plt.title('TOP  CRIME RATES PER ALCALDIA MEXICO CITY')
plt.xlabel('NUMBER OF CRIMES')
plt.ylabel('ALCALDIA')

   
    # Markers and colors as requested for the 4 treatments, the remaining at my choice.
x_axis = crimefinal2.index

count = 0

    #plot each tumor volume column with standard error
for c in crimefinal2.columns:
    sns.swarmplot(x_axis,crimefinal2[c],linestyle = '--',label=str(c))
    count += 1
#LEGEND AND FORMAT
plt.grid()
plt.tight_layout()
plt.rc('xtick', labelsize=6)

#plt.savefig('../Pymaceuticals/SEM_PLOT1.png')
plt.show()
```
